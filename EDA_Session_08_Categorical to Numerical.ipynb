**Do it for all columns at a time**

  - read the data

  - apply for loop

  - iterate through columns

  - get unique lables fro each column

  - get the count of unique labels

  - create a dictionary

  - applt in to map

```python
visa_df = pd.read_csv(r"C:\Users\Owner\OneDrive\Documents\NareshIT\DataScience\DataFiles\Visadataset - Visadataset.csv")
for col in visa_df.select_dtypes(include='object').columns:
    label = list(visa_df[col].unique())
    value = [i for i in range(len(label))]
    d = dict(zip(label,value))
    visa_df[col] = visa_df[col].map(d)
```

**One hot Encoder**

- one hot means, if one is ON another one is OFF

- ON represents with 1 OFF represents with 0

- for ex: case status has 2 unique labels
   - Denied
   - Certified
 
- If we apply one hot encoder on case status column, it will creates 2 extra columns
   - case_status_Denied
   - case_status_Certified

|case_status|case_status_denied|case_status_certified|
------|---|---|
|Denied|1|0|
|Certified|0|1|

**Advantages**

- In Machine Learning we have one important proprety is, **independent of variables**.

- Machine learning models excepts variables are independent to each other

- one hot encoder creates new columns, those columns are **independent to each other**

- Independent variables means both have **90 degree phase shift**

- 90 degree phase shift also called as **perpendicular to each other**

- perpendicular means **orthogonal to each other**

**Disadvantages**

- one hot encoder creates new columns equal to no. of unique labels

- if no. of unique labels increase, new columns also increases

- The more columns leads data is more

- when data is more, complexity increases

- with complexity increse more computation power will be required

- Also time increases

- The entire process is called **curse of dimentionality**

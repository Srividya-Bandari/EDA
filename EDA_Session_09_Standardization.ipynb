### Standardization

- Standardization is the technique, used to scale all the data under one scale

- different columns has different values with different units

- some column values has bigger values, some values has lesser values

- so it is important to keep all the values under one scale

- we have 2 methods
   - standardization
   - normalization
 
- Standardization
   - Z-Score
   - Z-score varies values from -3 to 3
 



- Normalization
   - min max scalar
   - value varies from -1 to 1
 


**task**

- step-1: take prevailing wage column

- step-2: calculate mean of prevailing wage column

- step-3: calculate standard deviation of prevailing wage column

- step-4: calculate the Nr:- step-1 - step-2

- step-5: divide step-4/step-3

```python
mean = visa_df['prevailing_wage'].mean()
std = visa_df['prevailing_wage'].std()
Nr = visa_df['prevailing_wage'] - mean
visa_df['new_prevailing_wage'] = Nr/std
print(visa_df['prevailing_wage'].idxmax(), visa_df['new_prevailing_wage'].idxmax())
print(visa_df['prevailing_wage'].idxmax(), visa_df['new_prevailing_wage'].idxmin())
```

**Normalization**

```python
visa_df = pd.read_csv(r"C:\Users\Owner\OneDrive\Documents\NareshIT\DataScience\DataFiles\Visadataset - Visadataset.csv")
x_min = visa_df['prevailing_wage'].min()
x_max = visa_df['prevailing_wage'].max()
Nr = visa_df['prevailing_wage'] - x_min
Dr = x_max - x_min
visa_df['norm_prevailing_wage'] = Nr/Dr
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
visa_df['mms_prevailing_wage'] = mms.fit_transform(visa_df[['prevailing_wage']])
```

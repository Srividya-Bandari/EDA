**Data transformatiom**

- Generally data has 3 types of distributions

   - Normal distribution
 
   - Positive skewed
 
   - Negative skewed
 
- All the maths concepts developed by assumption as Data follows Normal distribution

- Data is skewed because of many reasons, inthat one is Outliers

- Once we perform the outlier analysis then we can apply transformation methods, if data does not follows normality

- We have some transformation methods

   - Log transformation
 
   - Exponential transformation
 
   - Reciprocal transformation
 
   - Square transformation
 
   - Power transformation

$Step-1

- import packages

```python
import numpy as np
import matplotlib.pyplot as plt
```

$Step-2$
- read data

```python
exp_data = np.random.exponential(size = 10000)
exp_data
```

$Step-3$

- plot the data

```python
plt.hist(exp_data, bins=20,label='exponential data')
plt.show()
```

```python
norm_data = np.random.normal(size = 10000)
norm_data
```

```python
plt.hist(norm_data, bins=20,label='exponential data')
plt.show()
```

$Step-4$

- **Convert exponential data to normal data**

**Using Log Transformation**

- Log transformation means performing logarithm operation on original data

- it is an approach to convert to normal distribution

- log means naturl log, i.e. to the base e

```python
log_data = np.log(exp_data)
```

```python
plt.hist(log_data, bins=20,label='log data')
plt.show()
```

```python
plt.figure(figsize=(15,4))
plt.title("expo vs log")
plt.subplot(1,2,1).hist(exp_data, bins=20,label='exponential data')
plt.legend()
plt.subplot(1,2,2).hist(log_data, bins=20,label='log data')
plt.legend()
plt.show()
```

**Using Reciprocal Transformation**

- Reciprocal will fail when values are small

- for ex: x=0, 1/0 fails

```python
#rec_data = 1/exp_data
rec_data = np.reciprocal(exp_data)
```

```python
rec_data
```

```python
rec_data = 1/exp_data
rec_data
```

```python
plt.hist(rec_data, bins = 20, lable = 'reciprocal data')
plt.show()
```

**Squate root transformation**

```python
sq_data = np.sqrt(exp_data)
sq_data
```

```python
plt.hist(sq_data, bins=20)
plt.show()
```

```python
plt.figure(figsize=(15,4))
plt.title("expo vs log vs sqrt")
plt.subplot(1,3,1).hist(exp_data, bins=20,label='exponential data')
plt.legend()
plt.subplot(1,3,2).hist(log_data, bins=20,label='log data')
plt.legend()
plt.subplot(1,3,3).hist(sq_data, bins=20,label='sqrt data')
plt.legend()
plt.show()
```

**Power transformer**

- Box cox



- ye-jhonson

- It is under sklearn

```python
import pandas as pd
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer(method='box-cox')
pt_data = pt.fit_transform(pd.DataFrame(exp_data))
```

```python
plt.hist(pt_data, bins =20)
```

```python
plt.figure(figsize=(15,4))
plt.subplot(1,1,1).hist(exp_data, bins=20,label='exponential data')
plt.legend()
plt.subplot(1,2,2).hist(pt_data, bins=20,label='power data')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer() # method = default -> yhe jhonson
pt_data = pt.fit_transform(pd.DataFrame(exp_data))

plt.hist(pt_data, bins =20)
```

```python
plt.figure(figsize=(15,4))
plt.subplot(1,1,1).hist(exp_data, bins=20,label='exponential data')
plt.legend()
plt.subplot(1,2,2).hist(pt_data, bins=20,label='power data')
plt.legend()
plt.show()
```

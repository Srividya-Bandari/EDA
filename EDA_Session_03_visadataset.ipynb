**Task**

- create categorical column list
- create numerical column list

- idea:
- convert series to dictionary
- iterate through loop
- apply the condition get the results in the list


```python
d = dict(visa_df.dtypes)
d
```

```python
num = []
cat = []
for i in d:
    if d[i] == 'int64' or d[i] == 'float64':
        num.append(i)
    else:
        cat.append(i)

num,cat
```

```python
cat = [key for key, values in d.items() if values == 'object']
cat
```

```python
num = [key for key,values in d.items() if values != 'object']
num
```
